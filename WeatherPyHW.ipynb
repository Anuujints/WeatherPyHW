```python
# Setting up dependencies, and importing API key
import matplotlib.pyplot as plt
import requests
import pandas as pd
import numpy as np
import time

from api_key import api_keys
from citipy import citipy
```

```python
# Output File (CSV)
datas = "cities_data.csv"
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_and_long = []
cities = []

# Creating Lat and Long
lat = np.random.uniform(low=-90.000, high=90.000, size=2000)
long = np.random.uniform(low=-180.000, high=180.000, size=2000)
lat_and_long= zip(lat, long)

for lat_and_long in lat_and_long:
    city = citipy.nearest_city(lat_and_long[0], lat_and_long[1]).city_name
    
    if city not in cities:
        cities.append(city)

len(cities)

lat_range = (-90, 90)
lng_range = (-180, 180)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
api_key = api_keys


# List of city data
cities_data = []

print("Beginning Data Retrieval     ")
print("-----------------------------")

count_r = 1
count_s = 1

for i, city in enumerate(cities):
        
    # Creating URL and City URL
    url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key
    city_url = url + "&q=" + city


    if (i % 50 == 0 and i >= 50):
        count_s += 1
        count_r = 0    

    print("Processing Record %s of Set %s | %s" % (count_r, count_s, city))
    
    # Add 1 to the record count
    count_r += 1

    # Run an API request for each of the cities
    try:
        # Parse the JSON and retrieve data
        response = requests.get(city_url).json()

        # Parse out the max temp, humidity, and cloudiness
        cityLat = response["coord"]["lat"]
        cityLong = response["coord"]["lon"]
        max_temp = response["main"]["temp_max"]
        humidity = response["main"]["humidity"]
        clouds = response["clouds"]["all"]
        wind = response["wind"]["speed"]
        country = response["sys"]["country"]
        date = response["dt"]

        # Append the City information into city_data list
        cities_data.append({"City": city, 
                          "Lat": cityLat, 
                          "Long": cityLong, 
                          "Max Temp": max_temp,
                          "Humidity": humidity,
                          "Cloudiness": clouds,
                          "Wind Speed": wind,
                          "Country": country,
                          "Date": date})
        
      

    # If no record found "skip" to next call
    except:
        print("City not found. Skipping...")
        pass
              
# Indicate that Data Loading is complete 
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
        
print(cities_data)
```

```python
### Convert Raw Data to DataFrame
# Export the city data into a .csv.
city_df = pd.DataFrame(cities_data)

city_lt = city_df["Lat"]
temp_max = city_df["Max Temp"]
humidity = city_df["Humidity"]
cloudiness = city_df["Cloudiness"]
wind_speed = city_df["Wind Speed"]

city_df.to_csv('data.csv')
city_df.count()
```

# Latitude vs. Wind Plot

```python
# Build the scatter plots for latitude vs. wind speed
plt.scatter(city_lt, 
            wind_speed,
            edgecolor="black", linewidths= 1, marker="o", 
            alpha=1, label="Cities")

plt.title("City Latitude vs. Wind Speed (%s)" % time.strftime("%x"))
plt.ylabel("Wind Speed in mph")
plt.xlabel("Latitude")
plt.grid(True)

plt.savefig("latVsWind.png")
plt.show()
```
